**a. A good measure of distance between two instances with categorical features is the overlap metric (also known as the hamming distance), which simply counts the number of descriptive features that have different values. Using this measure of distance, compute the distances between the mystery animal and each of the animals in the animal dataset.**

```python
from collections import Counter


dataset = [
    {"id": 1, "class": "Mammal", "features": [True, False, True, True, False, False, True, False]},
    {"id": 2, "class": "Amphibian", "features": [False, True, False, False, True, True, False, False]},
    {"id": 3, "class": "Mammal", "features": [True, False, True, True, False, False, True, False]},
    {"id": 4, "class": "Bird", "features": [False, True, False, True, False, True, False, True]},
]


query_features = [False, True, False, False, False, True, False, False]


def hamming_distance(features1, features2):
    return sum(f1 != f2 for f1, f2 in zip(features1, features2))


distances = []
for animal in dataset:
    dist = hamming_distance(animal["features"], query_features)
    distances.append((dist, animal["class"]))
    print(f"Distance to {animal['class']} (ID {animal['id']}): {dist}")
```

**b. If you used a 1-NN model, what class would be assigned to the mystery animal?**

```python
distances.sort()
knn_1_class = distances[0][1]

print("\n1-NN Classification:", knn_1_class)
```

<strong>c. If you used a 4-NN model, what class would be assigned to the mystery animal? Would this be a good value for k for this dataset?</strong>

```python
k = 4
knn_4_classes = [cls for _, cls in distances[:k]]
most_common_class = Counter(knn_4_classes).most_common(1)[0][0]

print("4-NN Classification:", most_common_class)
```
